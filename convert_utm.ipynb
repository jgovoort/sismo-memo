# Convert Deg. into UTM

```python
import numpy as np
import pandas as pd
import utm
```

```python
def utm_converter(X, Y) :
    outX = []
    outY = []
    for i in range(len(X)) :
        UTM = utm.from_latlon(Y[i],X[i])
        outX.append(UTM[0]), outY.append(UTM[1])
    return outX, outY
```

```python
Tiago  = pd.read_csv('C:/Users/julie/Downloads/station_info_SoloP_buildingN.csv')
```

```python
Tiago
```

```python
BINlat = np.isfinite(Tiago.Latitude)
lat = np.extract(BINlat, Tiago.Latitude)
BINlong = np.isfinite(Tiago.Longitude)
long = np.extract(BINlat, Tiago.Longitude)
```

```python
lat, long
```

```python
utmLong, utmlat = utm_converter(long, lat)
```

```python
utmLong.extend([np.nan]*24)
utmlat.extend([np.nan]*24)
```

```python
Tiago.insert(8, 'Easting',utmLong, True)
Tiago.insert(8, 'Northing',utmlat, True)
```
